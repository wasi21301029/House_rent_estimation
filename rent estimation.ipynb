```python
#packege imports
from sklearn import preprocessing, linear_model
import numpy as np

import sklearn
```

```python
#Data reading
import pandas as pd

df = pd.read_csv("new_house_data.csv")

df.drop(["id"],1)
df.head()
```

```python
#Data processing

df["Area"] = df["Area"].map(lambda i:i[0:].replace(" sqft",""))



df["Area"] = df["Area"].map(lambda i:i[0:].replace(",",""))
df["Area"] = df["Area"].map(lambda i:int(i))


df["Price"] = df["Price"].map(lambda i:(i[0:].replace(" Thousand","000")))

df["Price"] = df["Price"].map(lambda i:(i[0:].replace(" Lakh","0000").replace(".","")))
df["Price"] = df["Price"].map(lambda i:int(i))


df.head()
```

```python
#Checking null value

df.isnull().sum()
```

```python
#Spliting Data
df = df.drop(["Location"],1)

x = np.array(df.drop(["Price"],1))
y = np.array(df["Price"])
print("x",x.shape)
print("y",y.shape)

xTrain,xTest, yTrain, yTest = sklearn.model_selection.train_test_split(x,y,
                                                                      test_size = 0.5,
                                                                      random_state = 10)
print(xTrain.shape)
print(xTest.shape)
```

```python
#model Training

model = linear_model.LinearRegression()
model.fit(xTrain,yTrain)
accuracy = model.score(xTest,yTest)

print("Coefficiants",model.coef_)
print("Y intercept",round(model.intercept_,3))

print("Accuracy",round(accuracy*100,3),"%")
```

```python
#Evaluation
testvals = model.predict(xTest)
#print(testvals)

error = []

for i,testval in enumerate(testvals):
    error.append(yTest[i]-testval)
    
    print(f"Actual value {yTest[i]}; prediction {int(testval)}; Error: {round(error[i])}")
    
```
