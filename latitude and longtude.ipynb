```python
#removing the "R-S" from location column

import pandas as pd


df = pd.read_csv("houserentdhaka_new.csv")

df["Location"] = df["Location"].map(lambda i:i[0:].replace(" R-A",""))

df.head()
```

```python
#Checking for known and unknown GeoCodes


import pandas as pd
import requests
import urllib.parse


def get_geo(address):
    address = address
    url = 'https://nominatim.openstreetmap.org/search/' + urllib.parse.quote(address) +'?format=json'

    response = requests.get(url).json()
    return response


locations = df["Location"]


known = []
unknown = []
for i in range(len(locations)):
    
    resp = get_geo(locations[i])
    
    if len(resp)==0:
        print("Unknown",i,locations[i])
        tup = (i,locations[i])
        unknown.append(tup)
    else:
        print("Known",i,locations[i])
        lat = resp[0]["lat"]
        long = resp[0]["lon"]
        
        tup = (i,lat,long)
        
        known.append(tup)
print(len(known))
print(len(unknown))

```

```python
#Ration of latidude and logntude
new_df = pd.read_csv("houses.csv")
new_df.head()

for tup in known:
    ratio = float(tup[1])/float(tup[2])
    
    new_df["lat-lon"][tup[0]] = ratio
    
new_df.head()
```
